#### 指数分布

```python
from __future__ import print_function, division

%matplotlib inline

import numpy as np

import nsfg
import first
import analytic

import thinkstats2
import thinkplot
```

```python
df = analytic.ReadBabyBoom()
diffs = df.minutes.diff()
#diff()を使うと差分、到着時刻差を表す
```

```python
cdf = thinkstats2.Cdf(diffs,label="actual")
thinkplot.Cdf(cdf)
thinkplot.Show(xlabel ="minute",ylabel="CDF")
```

```python
thinkplot.Cdf(cdf,complement=True)
thinkplot.Show(xlabel ="minute",ylabel="log(CDF)",yscale="log")
```

#### 正規分布

```python
import scipy.stats
scipy.stats.norm.cdf(0)
```

```python
thinkplot.PrePlot(3)

mus = [1.0, 2.0, 3.0]
sigmas = [0.5, 0.4, 0.3]
for mu, sigma in zip(mus, sigmas):
    xs,ps = thinkstats2.RenderNormalCdf(mu=mu, sigma=sigma, 
                                               low=-1.0, high=4.0)
    label = r'$\mu=%g$, $\sigma=%g$' % (mu, sigma)
    thinkplot.Plot(xs,ps, label=label)

thinkplot.Config(title='Normal CDF', xlabel='x', ylabel='CDF',
                 loc='upper left')
```

#### 乱数の生成

```python
import math
def abc(lam):
    p =np.random.random()
    x = -math.log(1-p)/lam
    return x
```

```python
abc(1)
```

```python
abc(1)
```

```python
t = [abc(lam=1) for _ in range(1000)] #1000回乱数を発生させる
```
